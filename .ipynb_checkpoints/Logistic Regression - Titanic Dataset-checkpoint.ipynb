```python
# importing modules that are requried in this notebook

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# loading the data 

data=pd.read_csv(r"D:\Titanic Dataset\Titanic Dataset.csv")
```

```python
# printing first 5 rows of data

data.head()
```

```python
# describing the data

data.describe()
```

```python
# finding the number of null values in each column

data.isnull().sum()
```

```python
# filling the 'Age' column null values with the mean of 'Age' column

data["Age"].fillna(data["Age"].mean(),inplace=True)
```

```python
# droping column that are not useful

# Note : dropping 'Cabin' column beacuse in dataset of 668 rows 
#        there are 514 null values 

data.drop(['Name','Ticket','Cabin','Embarked'],axis=1,inplace=True)
```

```python
# creating Sex_Num function which in 'Sex' column ,
# return 1 when Sex=='Male' else 2

def Sex_Num(gender):
    
    if(gender=='male'):
        return 1
    else:
        return 2
```

```python
# using Sex_Num function and storing it in 'sex' 
# dropping the origional 'Sex' Column

data['sex']=data['Sex'].apply(Sex_Num)

data.drop('Sex',axis=1,inplace=True)
```

```python
# printing first 5 rows of cleaned data

data.head()
```

```python
# storing input and target/output column

X=data.drop("Survived",axis=1)
Y=data["Survived"]
```

```python
# splitting the data into train and test data

X_train , X_test , Y_train , Y_test=train_test_split(X,Y) 
```

```python
# Scaling the features

scaler = StandardScaler()

scaler.fit(X_train)

X_train_transformed=scaler.transform(X_train)
X_test_transformed=scaler.transform(X_test)
```

```python
# applying the LogisticRegression on train data

alg=LogisticRegression()

alg.fit(X_train_transformed,Y_train)

# predicting the output for the test input data

prediction=alg.predict(X_test_transformed)

# printing the confusion matrix of test data

pd.DataFrame(confusion_matrix(Y_test,prediction),columns=["0_Predicted","1_Predicted"],
             index=["0_True","1_True"])
```

```python
# printing the classification_report of test data

print(classification_report(Y_test,prediction))
```
